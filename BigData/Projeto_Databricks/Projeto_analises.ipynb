```
%sql
use projeto_pmagno;
```

```
%sql
/* 1 - Quais são os três produtos MENOS vendidos?
*/
SELECT
      product_name,
      ROUND(SUM((unit_price * quantity * (1-discount))) + AVG(freight),2) AS valor_total
FROM ft_sale
GROUP BY 1
ORDER BY 2
LIMIT 3
```

```
%sql
/* 2 - Quais são os cinco clientes que mais compras fizeram?*/

SELECT
      company_customer,
      COUNT(order_id) as quant_purchase
FROM ft_sale
GROUP BY 1
ORDER BY 2 DESC
LIMIT 5
```

```
%sql
/* 3 - Quais são os cinco clientes com maior total de vendas? */

SELECT
      company_customer,
      ROUND(SUM((unit_price * quantity * (1-discount))) + AVG(freight),2) AS valor_total
FROM ft_sale
GROUP BY 1
ORDER BY 2 DESC
LIMIT 3
```

```
%sql
/* 4 - Qual o melhor funcionário do último mês registrado? (total de vendas) */

SELECT
      A.employee_first_name,
      A.employee_last_name,
      ROUND(SUM((unit_price * quantity * (1-discount))) + AVG(freight),2) AS valor_total,
      B.year_date,
      B.month_date
FROM ft_sale AS A
LEFT JOIN dim_time AS B ON A.order_date = B.calendar_id
GROUP BY 1,2,4,5
ORDER BY 4 DESC, 5 DESC, 3 DESC
LIMIT 1
```

```
%sql
/* 5 - Quais as regiões com menos clientes cadastrados? */

SELECT
      region_description,
      COUNT(company_customer) as quant_customers
FROM ft_sale
GROUP BY 1
ORDER BY 2 
```
