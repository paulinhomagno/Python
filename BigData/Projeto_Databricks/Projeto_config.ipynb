## Notebook de configuração dos caminhos dos diretórios das camadas

```
# caminho para a 1 camada (raw)
path_raw = f"/letscode/pmagno/projeto/raw"

# criando uma base de dados para armazenamento das tabelas
spark.sql(f"CREATE DATABASE IF NOT EXISTS projeto_pmagno")
spark.sql(f"USE projeto_pmagno")
```

```
# listas com os nomes das tabelas otimizar a criação de tabelas e diretórios nas camadas
list_tables = ['customers', 'employees', 'employee_territories', 'orders', 'order_details', 
           'products', 'categories','region', 'shippers', 'suppliers', 'territories', 'us_states']

# lista que receberá o caminho da 2 camada (trusted)
paths_trusted = []

# loop para a criação dos diretórios dentro da camada trusted conforme o nome das tabelas
for i in range (len(list_tables)):
    
    path = f"/letscode/pmagno/projeto/trusted/{list_tables[i]}"
    paths_trusted.append(path)
```
