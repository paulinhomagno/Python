##Tabelas dimensao

Criando dimensao tempo com datas da base de dados e inserindo ano, mes, dia e dia da semana

```
from pyspark.sql.functions import explode, sequence, to_date

# data mais antiga pelo order_date e mais recente pelo required_date

df = spark.read.table("orders_trusted")
beginDate =  df.agg({"order_date": "min"}).collect()[0][0]
beginDate = beginDate.strftime("%Y-%m-%d")
endDate = df.agg({"required_date": "max"}).collect()[0][0]
endDate = endDate.strftime("%Y-%m-%d")

# criando uma view com o intervalo de dias entre o dia mais antigo e o mais recente
(
  spark.sql(f"select explode(sequence(to_date('{beginDate}'), to_date('{endDate}'), interval 1 day)) as calendarDate")
    .createOrReplaceTempView('dates')
)
```

```
%sql

/* criando tabela dimensao tempo com a data completa e separando ano, mês, dia e dia da semana
*/

CREATE OR REPLACE TABLE dim_time
USING DELTA
LOCATION '/letscode/pmagno/projeto/refined/dimensions/dim_time'
AS SELECT
  ROW_NUMBER() OVER(ORDER BY calendarDate) AS calendar_id,
  calendarDate as calendar_date,
  year(calendarDate) as year_date,
  month(calendarDate) as month_date,
  day(calendarDate) as day_date,
  CASE dayofweek(calendarDate)
           WHEN 1 THEN 'Mon'
           WHEN 2 THEN 'Tues'
           WHEN 3 THEN 'Wed'
           WHEN 4 THEN 'Thu'
           WHEN 5 THEN 'Fri'
           WHEN 6 THEN 'Sat'
           WHEN 7 THEN 'Sun'
  END as weekday_date
FROM dates
```

```
%sql

/* criando a tabela dimensao orders  */

CREATE OR REPLACE TABLE dim_orders
USING DELTA
LOCATION '/letscode/pmagno/projeto/refined/dimensions/dim_orders'
SELECT A.order_id,
        A.employee_id,
        A.order_date,
        A.required_date,
        A.shipped_date,
        A.freight,
        A.customer_id,
        A.ship_country,
        A.ship_city,
        B.product_id,
        B.quantity,
        B.unit_price,
        B.discount
FROM orders_trusted AS A
RIGHT JOIN order_details_trusted AS B
ON A.order_id = B.order_id
```

```
%sql

/* criando a tabela dimensao products  */

CREATE OR REPLACE TABLE dim_products
USING DELTA
LOCATION '/letscode/pmagno/projeto/refined/dimensions/dim_products'
SELECT A.product_id,
        A.product_name,
        A.category_id,
        B.category_name,
        B.description as description_category,
        A.unit_price,
        C.supplier_id,
        C.company_name,
        C.country,
        A.units_in_stock,
        A.quantity_per_unit
FROM products_trusted AS A
RIGHT JOIN categories_trusted AS B
ON A.category_id = B.category_id
RIGHT JOIN suppliers_trusted AS C
ON A.supplier_id = C.supplier_id
```

```
%sql
/* criando a tabela dimensao customers  */

CREATE OR REPLACE TABLE dim_customers
USING DELTA
LOCATION '/letscode/pmagno/projeto/refined/dimensions/dim_customers'
SELECT A.customer_id,
        A.company_name as company_customer,
        A.city as city_customer,
        A.region as region_customer,
        A.country as country_customer
FROM customers_trusted AS A
```

```
%sql

/* criando a tabela dimensao territories para formar a dimensão employees com o territory e region junto aos employees */

CREATE OR REPLACE TABLE dim_territories
USING DELTA
LOCATION '/letscode/pmagno/projeto/refined/dimensions/dim_territories'
SELECT A.employee_id,
        A.territory_id,
        B.territory_description,
        B.region_id,
        C.region_description
FROM employee_territories_trusted AS A
RIGHT JOIN territories_trusted AS B
ON A.territory_id = B.territory_id
RIGHT JOIN region_trusted AS C
ON B.region_id = C.region_id

```

```
%sql

CREATE OR REPLACE TABLE dim_employees
USING DELTA
LOCATION '/letscode/pmagno/projeto/refined/dimensions/dim_employees'
SELECT DISTINCT A.employee_id,
        A.last_name,
        A.first_name,
        A.birth_date,
        A.hire_date,
        B.region_id,
        B.region_description
FROM employees_trusted AS A
LEFT JOIN dim_territories AS B
ON A.employee_id = B.employee_id
```

## Tabela fato

```
%sql
CREATE OR REPLACE TABLE ft_sale
USING DELTA
LOCATION '/letscode/pmagno/projeto/refined/facts/ft_sale'
SELECT 
        ROW_NUMBER() OVER(ORDER BY A.order_id, D.product_id) AS sale_id,
        A.order_id,
        B.calendar_id as order_date,
        A.freight,
        C.company_customer,
        C.country_customer,
        D.product_id,
        D.product_name,
        D.unit_price,
        A.quantity,
        A.discount,
        D.category_name,
        D.company_name as company_product,
        D.country as country_product,
        F.last_name as employee_last_name,
        F.first_name as employee_first_name,
        F.region_description
        
FROM dim_orders AS A
LEFT JOIN  dim_time AS B ON A.order_date = B.calendar_date
LEFT JOIN dim_customers AS C ON A.customer_id = C.customer_id
LEFT JOIN dim_products AS D ON A.product_id = D.product_id
LEFT JOIN dim_employees AS F ON A.employee_id = F.employee_id
```

## Tabelas Agregadas

```
from pyspark.sql.functions import avg, countDistinct, sum


df_sales = spark.read.table("ft_sale")

# tabela agregada com região, quantidade de pedidos (quant_orders), qauntidade de clientes (quant_customers) e soma das quantidades vendidas (quant_sales)
df_orders_region = (
  df_sales
  .groupby("region_description")
  .agg(countDistinct(col("order_id")).alias("quant_orders"),
       countDistinct(col("company_customer")).alias("quant_customers"),
       sum(col("quantity")).alias("quant_sales"))
)


(df_orders_region.write
 .format("delta")
 .mode("overwrite")
 .save("/letscode/pmagno/projeto/refined/agg/orders_region"))

spark.sql(f"""
    DROP TABLE IF EXISTS orders_region
    """)

spark.sql(f"""
    CREATE TABLE orders_region
    USING DELTA
    LOCATION "/letscode/pmagno/projeto/refined/agg/orders_region"
    """)
```

```
%sql
/* tabela agrega com quantidade de cada produto vendido por região separado pelo ano mais recente e os meses */

CREATE OR REPLACE TABLE last_year_products_region
USING DELTA
LOCATION '/letscode/pmagno/projeto/refined/agg/last_year_products_region'
SELECT 
      A.product_name,
      A.region_description,
      B.year_date ,
      B.month_date,
      ROUND(SUM((unit_price * quantity * (1-discount))) + AVG(freight),2) AS valor_total            
FROM ft_sale AS A  
LEFT JOIN dim_time AS B ON A.order_date = B.calendar_id
WHERE B.year_date = (SELECT YEAR(MAX(calendar_date)) FROM dim_time)
GROUP BY 1,2,3,4
```

```
%sql
/* tabela agregada com a quantidade de pedidos e valor de vendas dos funcionários no ano mais recente e separado por m*/

CREATE OR REPLACE TABLE last_year_employee_sales
USING DELTA
LOCATION '/letscode/pmagno/projeto/refined/agg/last_year_employee_sales'
SELECT 
      A.employee_last_name,
      A.employee_first_name,
      B.year_date ,
      B.month_date,
      COUNT(DISTINCT A.order_id) AS quant_orders,
      ROUND(SUM((unit_price * quantity * (1-discount))) + AVG(freight),2) AS valor_total           
FROM ft_sale AS A  
LEFT JOIN dim_time AS B ON A.order_date = B.calendar_id
WHERE B.year_date = (SELECT YEAR(MAX(calendar_date)) FROM dim_time)
GROUP BY 1,2,3,4
```
