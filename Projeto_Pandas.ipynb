## 1. O Brasil nas Olimpíadas

Vamos começar estudando o desempenho do nossos próprio país. Gere um DataFrame novo contendo apenas as informações sobre atletas brasileiros.

````python
# localizando os atletas do Brasil pela coluna NOC, percebemos depois de que Team pode aparecer Brazil - 1 para alguns atletas
atletas_br = atletas_ol.loc[atletas_ol['NOC'] == 'BRA']
display(atletas_br.sample(10))
atletas_br.info()
````

### Medalhistas

Vamos focar um pouco nos casos de sucesso do Brasil. Use o seu DataFrame anterior para filtrar apenas informações sobre **medalhistas** brasileiros. 

**DICA:** observe como a coluna ```Medal``` é representada quando o atleta não ganhou medalha.

````python
# o dataset possui muitos valores ausentes em altura e peso
# uma forma de preencher os valores faltantes, é colocar a média de altura do esporte pelo genero
atletas_br['Height'] = atletas_br['Height'].fillna(atletas_br.groupby(['Sex', 'Sport'])['Height'].transform('mean'))
# apenas um valor nao foi preenchido, pois possui somente um atleta de um esporte que nao esta mais em olimpiadas, será preenchido com zero na proxima celula
display(atletas_br[atletas_br['Height'].isna()])
````

````python
# preenchendo com 0 o valor ausente
atletas_br[atletas_br['Height'].isna()]=0
atletas_br.info()
````

````python
#atletas_br[atletas_br['Weight'].isna()]
# implementando a mesma resolução para o peso
atletas_br['Weight'] = atletas_br['Weight'].fillna(atletas_br.groupby(['Sex', 'Sport'])['Weight'].transform('mean'))
# veriicando se possui ainda dados ausentes
display(atletas_br[atletas_br['Weight'].isna()])
display(atletas_br.info())
````

````python
# verificando os valores ausentes em Idade, se percebe que são de edições mais antigas das olimpiadas
display(atletas_br[atletas_br['Age'].isna()])
````

````python
# para preencher esses valores, usams a mesma logica das anteriores, mas, como tentativa para se chegar na realidade daquela temporada
# usamos o media do esporte e do ano para preenche-los, ja que todos sao do sexo masculino
atletas_br['Age'] = atletas_br['Age'].fillna(atletas_br.groupby(['Sport', 'Year'])['Age'].transform('mean'))
display(atletas_br[atletas_br['Age'].isna()])
display(atletas_br.info())
````

````python
#ainda restaram valores ausentes, provalmente por nao exister outros atletas neste criterio para puxar os dados de média
# assim, usamos a logica de gerero e esporte para completar esses valores
atletas_br['Age'] = atletas_br['Age'].fillna(atletas_br.groupby(['Sex', 'Sport'])['Age'].transform('mean'))
display(atletas_br[atletas_br['Age'].isna()])
display(atletas_br.info())
````

````python
#  dataframe com atletas medalhistas usando dropna que irá retirar as linhas(atletas) com medal com valor nan
atletas_br_medal = atletas_br.dropna(subset=['Medal'])
atletas_br_medal
````

Será que os dados acima influenciaram no interesse geral dos atletas pelo esporte ou realmente impactaram no desempenho deles? Podemos tentar descobrir se há algum tipo de correlação.

Você ainda possui o dataframe original contendo todos os atletas brasileiros, incluindo os sem medalha? Obtenha os valores médios de peso e altura por esporte daquele dataframe e compare-o com os dos medalhistas. Há alguma diferença significativa em algum esporte?

````python
print('Altura média, atletas em geral:')
media_altura_geral = atletas_br.groupby('Sport')['Height'].mean()
display(media_altura_geral)
````

````python
print('Peso médio, atletas em geral:')
media_peso_geral = atletas_br.groupby('Sport')['Weight'].mean()
display(media_peso_geral)
````

````python
# dataframe com a diferenca de media do peso
diferenca_peso = media_peso_med - media_peso_geral

# preenchendo com 0 os valores ausentes, que no caso sao esportes onde o Brasil nao tem medalhas
print('Diferença da média de peso dos medalhistas pela media geral:')
display(diferenca_peso.fillna(0))
# esportes onde media de peso dos medalhistas é maior que a media geral
print('\n')
maior_peso = diferenca_peso[diferenca_peso >0]
print('Esportes onde a média dos medalhistas é maior e sua diferença da média geral:')
# muitos esportes a media de peso dos medalhistas sao maiores e significativos
# no taekwondo e ginastica sao as maiores diferenças para os medalhistas
display(maior_peso)
print('\n')

# esportes onde media peso dos medalhistas é menor que a media geral

print('Esportes onde a média geral é maior e sua diferença da média dos medalhistas:')
menor_peso = diferenca_peso[diferenca_peso <0]*-1
display(menor_peso)
````

Qual foi (ou quais foram) o maior medalhista brasileiro em quantidade total de medalhas?

````python
# ESTA É  PRIMEIRA SOLUÇÃO QUE VEIO, USANDO FOR
#maior_medal = atletas_br_medal.groupby('Name')['Medal'].size()#
#maior_medal.loc[maior_medal.idxmax()]##

# agrupando por nome e utilizando função size para somar as medalhas totais por nome
maior_medal = atletas_br_medal.groupby('Name')['Medal'].count().reset_index()

# atribuindo o valor maximo de medalhas totais
n_medalhas = maior_medal.max()

# lista vazia para receber atleta(s) com maior numero de medalhas
lista = []

# iterando sobre o DF agrupada e condicional para localizar os nomes que possuem o valor maximo de medalhas
for x in range (len(maior_medal)):
    #print(maior_medal.iloc[x,1])
    if maior_medal.iloc[x,1] == int(n_medalhas):
        lista.append(maior_medal.iloc[x,0])
        
print('Atleta(s) com maior numero de medalhas:', lista)
````

````python
# SEGUNDA SOLUÇÃO, MAIS SIMPLES, MAS DEPOIS DE MUITOS ERROS
maior_medal = atletas_br_medal.groupby('Name')['Medal'].size().reset_index()
display(maior_medal)

# atribuindo o valor maximo de medalhas totais
n_medalhas = maior_medal.max()
# localizando maiores medalhistas pelo loc e filtro
maior_medal.loc[(maior_medal['Medal'] == n_medalhas[0])]
````

E o(s) maior(es) em quantidade de medalhas de ouro?

````python
# usando filtro para separar os medalhistas de ouro e ja agrupando por total de medalhas de ouro
medalhas_ouro = atletas_br_medal.loc[atletas_br_medal['Medal'] =='Gold'].groupby('Name')['Medal'].count().reset_index()

#medalhistas_ouro = medalhas_ouro.groupby('Name')['Medal'].count().reset_index()
# buscando o valor maximo de medalhas de ouro de um atleta
maximo = medalhas_ouro['Medal'].max()
# localizando o(s) atletas com medalhas de ouro igual ao maximo
medalhas_ouro.loc[medalhas_ouro['Medal'] == maximo]
````

Qual esporte rendeu mais medalhas de ouro para o Brasil? E qual rendeu mais medalhas no total?

**DICA:** tome muito cuidado nessa análise: cada **evento esportivo** rende 1 medalha. Por exemplo, quando a equipe de futebol vence, isso é considerado 1 medalha, mesmo tendo cerca de 20 atletas medalhistas na equipe. 

````python
# agrupando por ano esporte e nome a quantidade de medalhas
medalhas_sport_t = atletas_br_medal.groupby(['Year', 'Sport', 'Name'])['Medal'].count().reset_index()

# retirando atraves do drop_duplicates quando o ano e o esporte são iguais, ficando assim somente uma medalha por esporte e ano
medalhas_sport_t = medalhas_sport_t.drop_duplicates(subset=['Year','Sport'])

#agrupando por esporte para somar as medalhas
medalhas_sport_t = medalhas_sport_t.groupby('Sport')['Medal'].count().reset_index()

# atribuindo a quantidade maxima de medalhas que um esporte tem
maximo_medal = medalhas_sport_t['Medal'].max()
maximo_medal
# localizando o esportes que possuem esta quantidade de medalhas
print('Esporte com maior quantidade de medalhas:')
medalhas_sport_t.loc[medalhas_sport_t['Medal'] == maximo_medal]
````

````python
# separando os atletas com medalha de ouro
medalhas_sport_ouro = atletas_br_medal.loc[atletas_br_medal['Medal'] == 'Gold']

# utilizando mesma lógica do total de medalhas, agrupando por ano, evento e nome
medalhas_sport_ouro = medalhas_sport_ouro.groupby(['Year', 'Sport', 'Name'])['Medal'].count().reset_index()

# retirando atletas com medalha de mesmo ano e evento
medalhas_sport_ouro = medalhas_sport_ouro.drop_duplicates(subset=['Year','Sport'])

# somando as medalhas de ouro por evento(categoria do esporte)
medalhas_sport_ouro = medalhas_sport_ouro.groupby('Sport')['Medal'].count().reset_index()

# atribuindo o valor maximo de medalhas de ouro
maximo_ouro = medalhas_sport_ouro['Medal'].max()

# localizando categoria de esporte com essa soma de medalhas
print('Esporte(s) com maior quantidade de medalhas de ouro:')
medalhas_sport_ouro.loc[medalhas_sport_ouro['Medal'] == maximo_ouro]
````

Cada "categoria" dentro de um esporte é considerado um evento. Por exemplo, dentro de "atletismo", temos uma competição de 100m masculina, uma de 100m feminino, um revezamento 4 x 100m masculino, um revezamento 4 x 100m feminino, uma competição de 400m masculino, uma de 400m feminino, uma maratona masculina, uma maratona feminina, e assim sucessivamente.

Sabendo disso, qual evento esportivo mais rendeu medalhas de ouro para o Brasil? E total de medalhas?

````python
# agrupando por ano, evento(categoria do esporte) e nome
medalhas_evento_t = atletas_br_medal.groupby(['Year', 'Event', 'Name'])['Medal'].count().reset_index()

# retirando atraves do drop_duplicates quando o ano e o evento são iguais, ficando assim somente uma medalha por evento e ano
medalhas_evento_t = medalhas_evento_t.drop_duplicates(subset=['Year','Event'])

#agrupando por esporte para somar as medalhas
medalhas_evento_t = medalhas_evento_t.groupby('Event')['Medal'].count().reset_index()

# atribuindo a quantidade maxima de medalhas que um esporte tem
maximo_medal = medalhas_evento_t['Medal'].max()
maximo_medal
# localizando o esportes que possuem esta quantidade de medalhas
print('Categoria de esporte com maior quantidade de medalhas:')
medalhas_evento_t.loc[medalhas_evento_t['Medal'] == maximo_medal]
````

````python
# separando os atletas com medalha de ouro
medalhas_evento_ouro = atletas_br_medal.loc[atletas_br_medal['Medal'] == 'Gold']

# utilizando mesma lógica do total de medalhas, agrupando por ano, evento e nome
medalhas_evento_ouro = medalhas_evento_ouro.groupby(['Year', 'Event', 'Name'])['Medal'].count().reset_index()

# retirando atletas com medalha de mesmo ano e evento
medalhas_evento_ouro = medalhas_evento_ouro.drop_duplicates(subset=['Year','Event'])

# somando as medalhas de ouro por evento(categoria do esporte)
medalhas_evento_ouro = medalhas_evento_ouro.groupby('Event')['Medal'].count().reset_index()

# atribuindo o valor maximo de medalhas de ouro
maximo_ouro = medalhas_evento_ouro['Medal'].max()

# localizando categoria de esporte com essa soma de medalhas
print('Categoria de esporte(s) com maior quantidade de medalhas de ouro:')
medalhas_evento_ouro.loc[medalhas_evento_ouro['Medal'] == maximo_ouro]
````

Para finalizar sobre o Brasil: obtenha o total de medalhas de ouro, prata, bronze e total por ano.

````python
# agrupando as colunas necessarias para a contagem por tipo de medalhas
quantd_medalhas = atletas_br_medal.groupby(['Year','Event', 'Sex', 'Name'])['Medal'].value_counts().reset_index(name='Quantidade')

# se o ano e evento sõo iguais é removido, para contar somente uma medalha por evento   
quantd_medalhas = quantd_medalhas.drop_duplicates(subset=['Year','Event', 'Medal', 'Sex'])

# Agrupando por ano e quantidade de cada tipo de medalha
display(quantd_medalhas.groupby('Year')['Medal'].value_counts())

````

Qual país foi o maior medalhista de ouro da história das olimpíadas de verão? Lembre-se da questão do evento esportivo, para não considerar múltiplas medalhas para um mesmo evento (ex: uma equipe de futebol fazendo parecer que mais de 20 medalhas foram distribuídas).

````python
# foi agrupado por ano, pais, evento e atleta para que após possa retirar atletas de mesmo evento, ano e pais e relizar a contagem de medalhas
separar_ouro = atletas_ol.loc[atletas_ol['Medal'] == 'Gold'].groupby(['NOC','Year','Event','Name'])['Medal'].count().reset_index()
separar_ouro.drop_duplicates(subset=['NOC','Year', 'Event'], inplace=True)

paises_ouro = separar_ouro.groupby('NOC')['Medal'].count().reset_index()
max_ouro = paises_ouro['Medal'].max()
print('País(es) com maior numero de medalhas de ouro:', end=' ')
print(','.join(paises_ouro.loc[paises_ouro['Medal']==max_ouro]['NOC'].tolist()),'com', max_ouro, 'medalhas!' )
````

Qual país foi o maior medalhista em medalhas totais na história das olimpíadas de verão?

````python
# agrupando por pais, ano evento e atleta para após remover os atletas com tipo de medalha igual do mesmo pais no mesmo ano e evento
# e resetando o indice, colocado a ultima coluna como qtde dos tipos de medalhas
separar_medal = atletas_ol.groupby(['NOC','Year','Event','Name'])['Medal'].value_counts().reset_index(name = 'Quantidade')
separar_medal.drop_duplicates(subset=['NOC','Year', 'Event', 'Medal'], inplace=True)
# contando a qtde de medalhas por país
paises_qtd_medal = separar_medal.groupby('NOC')['Medal'].count().reset_index()

maior_medal_pais = paises_qtd_medal['Medal'].max()

print('País(es) com maior numero de medalhas:', end=' ')
print(','.join(paises_qtd_medal.loc[paises_qtd_medal['Medal']==maior_medal_pais]['NOC'].tolist()),'com', maior_medal_pais, 'medalhas!' )
````

Obtenha o total de medalhas de ouro, prata e total por edição das Olimpíadas de verão. Lembre-se da questão do evento esportivo.

````python
# utilizado DF anterior e contando por tipo de medalhas por ano
separado = separar_medal.groupby('Year')['Medal'].value_counts().reset_index(name='Quantidade')
display(separado.sample(5))
# criando pivot para colocar os tipos de medalhas como colunas e ano como índice
medalhasEdicao = separado.pivot(index='Year', columns='Medal', values='Quantidade')
# adicionando coluna a soma das medalhas
medalhasEdicao['Total'] = medalhasEdicao.sum(axis=1)
# atribuindo a ordem e as colunas a mostar conforme enunciando
mostrar = ['Gold', 'Silver', 'Total']
print('Ouro, prata e total')
medalhasEdicao[mostrar].sample(5)
````

## 3. Brasil vs Mundo

Para finalizar, vamos fazer algumas comparações entre Brasil e mundo. Qual o ranking do Brasil em cada edição das olimpíadas? Lembrando que o ranking é ordenado por medalhas de ouro.

````python
separar = atletas_ol.groupby(['NOC','Year','Event','Name'])['Medal'].value_counts().reset_index(name = 'Quantidade')
separar.drop_duplicates(subset=['NOC','Year', 'Event', 'Medal'], inplace=True)
# contando a qtde de medalhas por país
paises_medal = separar.groupby(['Year', 'NOC'])['Medal'].value_counts().reset_index(name = 'Quantidade')

# colocando os tipos de medalhas como colunas
paises_medal['Gold'] = paises_medal.loc[(paises_medal['Medal'] == 'Gold'), 'Quantidade']
paises_medal['Silver'] = paises_medal.loc[(paises_medal['Medal'] == 'Silver'), 'Quantidade'] 
paises_medal['Bronze'] = paises_medal.loc[(paises_medal['Medal'] == 'Bronze'), 'Quantidade'] 

# eliminando as colunas nao mais interessantes, devido ja ter a qtde de medalhas na colunas criadas acima
paises_medal.drop(['Medal', 'Quantidade'], axis=1, inplace = True)

# preenchendo com 0 onde nao tem medalhas
paises_medal.fillna(0, inplace = True)

# ordenando a listagem por ano conforme é feito nas olimpiadas, o paises aparecerão 3 vezes por ano 
#ja que existem 3 tipos de medalhas
paises_medal.sort_values(by = ['Year', 'Gold', 'Silver','Bronze'], ascending = [True, False, False, False], inplace = True)

# retirando paises duplicados no mesmo ano
agrupar = paises_medal.drop_duplicates(subset = ['Year', 'NOC'], keep = 'first').reset_index(drop = True)

# criando uma coluna com numero 1 para depois fazer uma soma acumulativa por ano para criar o ranking por ano
agrupar['Numero'] = 1
agrupar['Ranking'] = agrupar.groupby(['Year'])['Numero'].cumsum()

# filtrando por BRA e verificar o ranking por ano
agrupar.drop(['Gold','Silver','Bronze', 'Numero'], axis = 1, inplace = True)
print('Ranking do Brasil por olimpiadas:')
agrupar[agrupar['NOC']=='BRA']
````
